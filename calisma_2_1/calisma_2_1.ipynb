```python
import cv2
import numpy
import matplotlib.pylab as plt
%matplotlib inline
```

```python
img=cv2.imread('racecar-j.jpg')
img.shape
```

```python
# Okuduumuz grnty ekranda gsteriyoruz
plt.axis("on")
plt.imshow(img)
plt.show()
```

```python
swapped_image = numpy.zeros_like(img)
x_ust=1000
x_alt=750
y_ust=580
y_alt=435
parcalar=[]
for i in range(0,4):
    for j in range (0,4):
        parcalar.append(img[y_alt:y_ust,x_alt:x_ust,:])
        swapped_image[y_alt:y_ust,x_alt:x_ust,:]=img[y_alt:y_ust,x_alt:x_ust,:]
        x_ust=x_alt
        x_alt=x_alt-250
    x_ust=1000
    x_alt=750
    y_ust=y_alt
    y_alt=y_alt-145


x_ust=250
x_alt=0
y_ust=145
y_alt=0

p=0
for i in range(0,4):
    for j in range (0,4):
        swapped_image[y_alt:y_ust,x_alt:x_ust,:]=parcalar[p]
        p=p+1
        x_alt=x_ust
        x_ust=x_ust+250
    x_ust=250
    x_alt=0
    y_alt=y_ust
    y_ust=y_ust+145

plt.imshow(swapped_image)
plt.show()
cv2.imwrite('swapped_img.jpg',swapped_image)
```
